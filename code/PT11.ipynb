## keras에서 IMDB 가져오기

```python
# 영화평 데이터셋 imdb 로드
from keras.datasets import imdb

# num_words: 상위 10,000개의 단어만을 사용 지정
(train_data, train_labels), (test_data, test_labels) = imdb.load_data(num_words=10000)
```

```python
print(train_data.shape, train_labels.shape)
print(test_data.shape, test_labels.shape)
```

```python
train_data
```

### imdb 데이터셋 구조

```python
# 학습데이터의 첫 번째 리뷰의 정답과 리뷰 문장(단어 정수)
print(train_labels[0])  # 부정: 0 긍정: 1
print(train_data[0][:10])  # 앞 단어 10개(단어 정수) 출력
print(train_data[0][-10:])  # 뒤 단어 10개(단어 정수) 출력

# 테스트데이터의 첫 번째 리뷰의 정답과 리뷰 문장(단어 정수)
print(test_labels[0])  # 부정: 0 긍정: 1
print(test_data[0][:10])  # 앞 단어 10개(단어 정수) 출력
print(test_data[0][-10:])  # 뒤 단어 10개(단어 정수) 출력
```

### 리뷰 길이 분포 확인

```python
import numpy as np

review_lengths = [len(x) for x in train_data]
print("최대 길이:", np.max(review_lengths))
print("최소 길이:", np.min(review_lengths))
print("평균 길이:", np.mean(review_lengths))
print("중앙값 길이:", np.median(review_lengths))
```

### 레이블 분포 확인

```python
import collections

label_counts = collections.Counter(train_labels)
print("부정 리뷰 개수:", label_counts[0])
print("긍정 리뷰 개수:", label_counts[1])
```
